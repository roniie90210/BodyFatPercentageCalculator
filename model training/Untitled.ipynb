```python
import pandas as pd
import numpy as np
from sklearn.linear_model import LinearRegression
```

```python
data=pd.read_csv('bodyfat.csv')
data
```

```python
columns=data.columns.to_list()
```

```python
columns.remove('BodyFat')
```

```python
columns
```

```python
y=data['BodyFat']
X=data[columns]
```

```python
X_arr=X.values
y_arr=y.values
```

```python
lm=LinearRegression()
model=lm.fit(X_arr,y_arr)
```

```python
lm.predict([[1.0704,25,176,72.5,37.8,99.6,88.5,97.1,60,39.4,23.2,30.5,29,18.8]])
```

```python
import pickle

pickle.dump(model,open('model.pkl','wb'))
```
